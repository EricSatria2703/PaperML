```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense, SpatialDropout1D, Conv1D, MaxPooling1D, Bidirectional
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score

# Load dataset
sundanese = pd.read_csv('/content/drive/MyDrive/eksperimen/newdataset.csv')

# Extract features and labels
fitur = sundanese.iloc[:, 1].values
labels = sundanese.iloc[:, 0].values

# Encode labels
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)
labels_categorical = to_categorical(labels_encoded)

# Print label encoding mapping
print("Label encoding mapping:")
for index, label in enumerate(label_encoder.classes_):
    print(f"{index} -> {label}")

# Tokenize and pad sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(fitur)
sequences = tokenizer.texts_to_sequences(fitur)
word_index = tokenizer.word_index
max_len = max([len(x) for x in sequences])
X = pad_sequences(sequences, maxlen=max_len)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, labels_categorical, train_size=0.8, random_state=0)

# Training parameters
epochs = 10
batch_size = 256

# Function to build and train a model
def build_and_train_model(model_type='RNN'):
    model = Sequential()
    model.add(Embedding(len(word_index) + 1, 100, input_length=max_len))
    model.add(SpatialDropout1D(0.2))
    if model_type == 'RNN':
        model.add(SimpleRNN(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'LSTM':
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'CNN_LSTM':
        model.add(Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'))
        model.add(MaxPooling1D(pool_size=2))
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'RNN_LSTM':
        model.add(SimpleRNN(100, dropout=0.2, recurrent_dropout=0.2, return_sequences=True))
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(y_train.shape[1], activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=2)
    return model, history

# Train RNN model
rnn_model, rnn_history = build_and_train_model('RNN')

# Train LSTM model
lstm_model, lstm_history = build_and_train_model('LSTM')

# Train CNN-LSTM model
cnn_lstm_model, cnn_lstm_history = build_and_train_model('CNN_LSTM')

# Train RNN-LSTM model
rnn_lstm_model, rnn_lstm_history = build_and_train_model('RNN_LSTM')

# Plot training & validation accuracy values for RNN
plt.figure(figsize=(12, 6))
plt.plot(rnn_history.history['accuracy'], label='Train Accuracy (RNN)')
plt.plot(rnn_history.history['val_accuracy'], label='Validation Accuracy (RNN)')
plt.title('RNN Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for LSTM
plt.figure(figsize=(12, 6))
plt.plot(lstm_history.history['accuracy'], label='Train Accuracy (LSTM)')
plt.plot(lstm_history.history['val_accuracy'], label='Validation Accuracy (LSTM)')
plt.title('LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for CNN-LSTM
plt.figure(figsize=(12, 6))
plt.plot(cnn_lstm_history.history['accuracy'], label='Train Accuracy (CNN-LSTM)')
plt.plot(cnn_lstm_history.history['val_accuracy'], label='Validation Accuracy (CNN-LSTM)')
plt.title('CNN-LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for RNN-LSTM
plt.figure(figsize=(12, 6))
plt.plot(rnn_lstm_history.history['accuracy'], label='Train Accuracy (RNN-LSTM)')
plt.plot(rnn_lstm_history.history['val_accuracy'], label='Validation Accuracy (RNN-LSTM)')
plt.title('RNN-LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Function to evaluate the model
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_classes = y_pred.argmax(axis=-1)
    y_true = y_test.argmax(axis=-1)

    print(confusion_matrix(y_true, y_pred_classes))
    print(classification_report(y_true, y_pred_classes))
    accuracy = accuracy_score(y_true, y_pred_classes)
    precision = precision_score(y_true, y_pred_classes, average='weighted')
    recall = recall_score(y_true, y_pred_classes, average='weighted')

    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")

    # Calculate specificity for each class
    cm = confusion_matrix(y_true, y_pred_classes)
    specificity = []
    for i in range(cm.shape[0]):
        tn = np.sum(np.delete(np.delete(cm, i, axis=0), i, axis=1))
        fp = np.sum(np.delete(cm, i, axis=0)[:, i])
        specificity.append(tn / (tn + fp))
    avg_specificity = np.mean(specificity)
    print(f"Specificity: {avg_specificity}")

# Evaluate RNN model
print("RNN Model Performance")
evaluate_model(rnn_model, X_test, y_test)

# Evaluate LSTM model
print("LSTM Model Performance")
evaluate_model(lstm_model, X_test, y_test)

# Evaluate CNN-LSTM model
print("CNN-LSTM Model Performance")
evaluate_model(cnn_lstm_model, X_test, y_test)

# Evaluate RNN-LSTM model
print("RNN-LSTM Model Performance")
evaluate_model(rnn_lstm_model, X_test, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense, SpatialDropout1D, Conv1D, MaxPooling1D, Reshape, Permute
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load dataset
sundanese = pd.read_csv('/content/drive/MyDrive/eksperimen/newdataset.csv')

# Extract features and labels
fitur = sundanese.iloc[:, 1].values
labels = sundanese.iloc[:, 0].values

# Encode labels
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)
labels_categorical = to_categorical(labels_encoded)

# Print label encoding mapping
print("Label encoding mapping:")
for index, label in enumerate(label_encoder.classes_):
    print(f"{index} -> {label}")

# Tokenize and pad sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(fitur)
sequences = tokenizer.texts_to_sequences(fitur)
word_index = tokenizer.word_index
max_len = max([len(x) for x in sequences])
X = pad_sequences(sequences, maxlen=max_len)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, labels_categorical, train_size=0.8, random_state=0)

# Training parameters
epochs = 10
batch_size = 32

# Function to build and train a model
def build_and_train_model(model_type='RNN'):
    model = Sequential()
    model.add(Embedding(len(word_index) + 1, 100, input_length=max_len))
    model.add(SpatialDropout1D(0.2))
    model.add(Permute((2, 1)))
    if model_type == 'RNN':
        model.add(SimpleRNN(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'LSTM':
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'CNN_LSTM':
        model.add(Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'))
        model.add(MaxPooling1D(pool_size=2))
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    elif model_type == 'RNN_LSTM':
        model.add(SimpleRNN(100, dropout=0.2, recurrent_dropout=0.2, return_sequences=True))
        model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(y_train.shape[1], activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=2)
    return model, history

# Train RNN model
rnn_model, rnn_history = build_and_train_model('RNN')

# Train LSTM model
lstm_model, lstm_history = build_and_train_model('LSTM')

# Train CNN-LSTM model
cnn_lstm_model, cnn_lstm_history = build_and_train_model('CNN_LSTM')

# Train RNN-LSTM model
rnn_lstm_model, rnn_lstm_history = build_and_train_model('RNN_LSTM')

# Plot training & validation accuracy values for RNN
plt.figure(figsize=(12, 6))
plt.plot(rnn_history.history['accuracy'], label='Train Accuracy (RNN)')
plt.plot(rnn_history.history['val_accuracy'], label='Validation Accuracy (RNN)')
plt.title('RNN Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for LSTM
plt.figure(figsize=(12, 6))
plt.plot(lstm_history.history['accuracy'], label='Train Accuracy (LSTM)')
plt.plot(lstm_history.history['val_accuracy'], label='Validation Accuracy (LSTM)')
plt.title('LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for CNN-LSTM
plt.figure(figsize=(12, 6))
plt.plot(cnn_lstm_history.history['accuracy'], label='Train Accuracy (CNN-LSTM)')
plt.plot(cnn_lstm_history.history['val_accuracy'], label='Validation Accuracy (CNN-LSTM)')
plt.title('CNN-LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation accuracy values for RNN-LSTM
plt.figure(figsize=(12, 6))
plt.plot(rnn_lstm_history.history['accuracy'], label='Train Accuracy (RNN-LSTM)')
plt.plot(rnn_lstm_history.history['val_accuracy'], label='Validation Accuracy (RNN-LSTM)')
plt.title('RNN-LSTM Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Evaluate RNN model
rnn_y_pred = rnn_model.predict(X_test)
rnn_y_pred_classes = rnn_y_pred.argmax(axis=-1)
y_true = y_test.argmax(axis=-1)
print("RNN Model Performance")
print(confusion_matrix(y_true, rnn_y_pred_classes))
print(classification_report(y_true, rnn_y_pred_classes))
print(accuracy_score(y_true, rnn_y_pred_classes))

# Evaluate LSTM model
lstm_y_pred = lstm_model.predict(X_test)
lstm_y_pred_classes = lstm_y_pred.argmax(axis=-1)
print("LSTM Model Performance")
print(confusion_matrix(y_true, lstm_y_pred_classes))
print(classification_report(y_true, lstm_y_pred_classes))
print(accuracy_score(y_true, lstm_y_pred_classes))

# Evaluate CNN-LSTM model
cnn_lstm_y_pred = cnn_lstm_model.predict(X_test)
cnn_lstm_y_pred_classes = cnn_lstm_y_pred.argmax(axis=-1)
print("CNN-LSTM Model Performance")
print(confusion_matrix(y_true, cnn_lstm_y_pred_classes))
print(classification_report(y_true, cnn_lstm_y_pred_classes))
print(accuracy_score(y_true, cnn_lstm_y_pred_classes))

# Evaluate RNN-LSTM model
rnn_lstm_y_pred = rnn_lstm_model.predict(X_test)
rnn_lstm_y_pred_classes = rnn_lstm_y_pred.argmax(axis=-1)
print("RNN-LSTM Model Performance")
print(confusion_matrix(y_true, rnn_lstm_y_pred_classes))
print(classification_report(y_true, rnn_lstm_y_pred_classes))
print(accuracy_score(y_true, rnn_lstm_y_pred_classes))
```
